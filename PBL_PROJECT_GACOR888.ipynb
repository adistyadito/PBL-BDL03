```python
num_col = ['HSC', 'SSC', 'Computer', 'English', 'Semester', 'Last', 'Overall' ]
cat_col = ['Department', 'Gender', 'Income', 'Hometown', 'Preparation', 'Gaming',	'Attendance',	'Job',	'Extra']
```

```python
#Univariate Analysis

# Histogram untuk memvisualisasikan skewness
plt.figure(figsize=(12, 48))
for i, col in enumerate(num_col, 1):
    plt.subplot(len(num_col), 1, i)
    sns.histplot(data[col], bins=20, kde=True)
    plt.title(f'Histogram of {col} with KDE')
plt.tight_layout()
plt.show()
```

```python
# Pastikan kolom numerik valid (konversi ke float jika perlu)
for col in num_col:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Cek missing values di kolom numerik
print("Jumlah Missing Values:")
print(data[num_col].isnull().sum())

# Buang baris dengan NaN (jika ada)
df_clean = data[num_col].dropna()

# Cek apakah data kosong setelah dropna
if df_clean.empty:
    print("Semua data di kolom numerik kosong atau hanya NaN.")
else:
    # Visualisasi Boxplot
    plt.figure(figsize=(30, 15))
    sns.boxplot(data=df_clean, orient='h')
    plt.title('Boxplot of Numerical Variables')
    plt.show()
```

HSC, SSC, computer, english, last, dan overall memiliki beberapa outliers di sisi kiri yang menandakan bahwa ada nilai yang sangat rendah dibandingkan rentang data lainnya. sedangkan untuk kolom semester memiliki distribusi yang sangat lebar dibanding yang lain, menandakan bahwa ada variasi yang signifikan dalam data semester

```python
# melihat jumlah outlier dari setiap kolom numerikal dengan menggunakan IQR
outlier_counts = {}

# Menghitung IQR dan jumlah outlier untuk setiap kolom numerik
for col in num_col:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Hitung jumlah outlier
    outlier_count = data[(data[col] < lower_bound) | (data[col] > upper_bound)].shape[0]

    # Simpan hasil ke dalam dictionary
    outlier_counts[col] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': outlier_count
    }

# Mengkonversi hasil ke dalam DataFrame untuk lebih mudah dibaca
outlier_summary = pd.DataFrame(outlier_counts).T
print(outlier_summary)
```

dengan jumlah outlier dari masing masing numerikal kolom lebih sedikit dari jumlah datanya. oleh karena itu, kami melakukan penanganan outlier yaitu menggunakan winsorization (memangkas nilai outlier pada batas tertentu) untuk menangani outlier pada numerical column

```python
import pandas as pd
from scipy.stats.mstats import winsorize

# Winsorize setiap kolom numerik dalam num_col
for col in num_col:  # Iterasi pada list num_col
    data[col] = winsorize(data[col], limits=[0.05, 0.05])  # 5% trimming di kedua sisi

# Menampilkan deskripsi statistik setelah winsorization
print(data[num_col].describe())
```

```python
# Boxplot untuk memvisualisasikan setelah di winsorization
plt.figure(figsize=(12, 6))

# Select the columns correctly using a list
sns.boxplot(data=data[num_col], orient='h')

plt.title('Boxplot of Numerical Variables')
plt.show()
```

jika dilihat dari box plot untuk data yang sudah dilakukan winsorization bahwa kolom seperti HSC, english, semester, last, dan overall distribusinya sudah lebih stabil. namun kolom seperti SSC dan computer masih memiliki outlier sehingga belum sepenuhnya tertangani oleh metode winsorization. oleh karena itu, perlu dilakukan cek IQR kembali

```python
# Inisialisasi dictionary untuk menyimpan hasil
outlier_counts = {}

# Daftar kolom yang ingin dianalisis
selected_columns = ['SSC', 'Computer']

# Menghitung IQR dan jumlah outlier untuk kolom SSC dan Last
for col in selected_columns:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Hitung jumlah outlier
    outlier_count = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)].shape[0]

    # Simpan hasil ke dalam dictionary
    outlier_counts[col] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': outlier_count
    }

# Mengkonversi hasil ke dalam DataFrame untuk lebih mudah dibaca
outlier_summary = pd.DataFrame(outlier_counts).T
print(outlier_summary)
```

dari hasil IQR terlihat kolom ssc dan Computer masih memiliki outlier, maka kami lakukan median replacement pada kedua kolom tersebut

```python
import numpy as np

# List kolom yang masih memiliki outlier
outlier_columns = ['SSC', 'Computer']

# Fungsi untuk mengganti outlier dengan median
def replace_outliers_with_median(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        median = data[col].median()

        # Ganti nilai outlier dengan median
        data[col] = np.where((data[col] < lower_bound) | (data[col] > upper_bound), median, data[col])
    return data

# Terapkan median replacement hanya pada kolom tertentu
data_cleaned = replace_outliers_with_median(data, outlier_columns)

# Tampilkan deskripsi data setelah median replacement
print(data_cleaned[outlier_columns].describe())
```

```python
# Inisialisasi dictionary untuk menyimpan hasil
outlier_counts = {}

# Daftar kolom yang ingin dianalisis
selected_columns = ['SSC', 'Computer']

# Menghitung IQR dan jumlah outlier untuk kolom SSC dan Last
for col in selected_columns:
    Q1 = data_cleaned[col].quantile(0.25)
    Q3 = data_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Hitung jumlah outlier
    outlier_count = data_cleaned[(data_cleaned[col] < lower_bound) | (data_cleaned[col] > upper_bound)].shape[0]

    # Simpan hasil ke dalam dictionary
    outlier_counts[col] = {
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': outlier_count
    }

# Mengkonversi hasil ke dalam DataFrame untuk lebih mudah dibaca
outlier_summary = pd.DataFrame(outlier_counts).T
print(outlier_summary)
```
